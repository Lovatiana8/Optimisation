# Question 1 : Comment satisfaire les demandes des clients finaux tout en minimisant les coûts d'expédition.

# Variables de décision
-> Les quantités de produit transportées de l'usine vers les dépots, du dépôt vers les clients et de l'usine vers les clients:

- Xud : Quantité expédiée de l'usine u : (Liverpool, Brighton) au dépôt d : (Newcastle, Birmingham, London, Exeter).
- Ydc : Quantité expédiée du dépôt d : (Newcastle, Birmingham, London, Exeter) au clients c : (c1, c2, c3, c4, c5, c6).
- Zuc : Quantité expédiée de l'usine u : (Newcastle, Birmingham, London, Exeter) au clients c : (c1, c2, c3, c4, c5, c6).

# Fonction objectif

L'objectif est de minimiser les coûts totaux d'expédition, qui est la somme de trois types de coûts :

$$
T = \sum_{u} \sum_{d} C_{ud} \cdot X_{ud}
+ \sum_{d} \sum_{c} C_{dc} \cdot Y_{dc}
+ \sum_{u} \sum_{c} C_{uc} \cdot Z_{uc}
$$


$X_{ud}$ : Quantité de produits expédiés de l'usine u au dépôt d.\
$Y_{dc}$ : Quantité de produits expédiés du dépôt d au client c.\
$Z_{uc}$ :  Quantité de produits expédiés directement de l'usine u au client c.\
$C_{ud}$ : Coût unitaire d'expédition du produit de l'usine u au dépôt d.\
$C_{dc}$ : Coût unitaire d'expédition du produit de l'usine u au client c.\
$C_{uc}$ : Coût unitaire d'expédition du produit directement de l'usine u au client c.\
T : coûts totaux d'expédition

# Contraintes

## Contraintes de satisfaction de la demande des clients :

La somme des produits reçus par chaque client doit être égale à la demande de ce client.

$$
\sum_{d} Y_{dc} + \sum_{u} Z_{uc} = \text{Demande}_c \quad \forall c \in \{C1, C2, C3, C4, C5, C6\}
$$

## Contraintes de capacité des usines :

La somme des produits expédiés depuis chaque usine ne doit pas dépasser sa capacité de production maximale.

$$
\sum_{d} X_{ud} + \sum_{c} Z_{uc} \leq \text{Capacité}_u \quad \forall u \in \{\text{Liverpool, Brighton}\}
$$

## Contraintes de débit des dépôts :

La somme des produits transitant par chaque dépôt ne doit pas dépasser son débit maximal.

$$
\sum_{c} Y_{dc} \leq \text{Débit}_d \quad \forall d \in \{\text{Newcastle, Birmingham, London, Exeter}\}
$$

## Contraintes de flux des dépôts :

Le flux de produit entrant dans un dépôt doit être égal au flux sortant vers les clients.

$$
\sum_{u} X_{ud} = \sum_{c} Y_{dc} \quad \forall d \in \{\text{Newcastle, Birmingham, London, Exeter}\}
$$

## Contraintes d'acheminement :

les valeurs nulles ou absentes dans le tableau des coûts sont = 0.

$$
\text{Si } C_{ud} = -, \text{ alors } X_{ud} = 0
$$

$$
\text{Si } C_{dc} = -, \text{ alors } Y_{dc} = 0
$$

$$
\text{Si } C_{uc} = -, \text{ alors } Z_{uc} = 0
$$


Pour résoudre ce problème d'optimisation, nous allons utiliser PuLP.

```python
# ! pip install pulp
import pulp
```

```python
# Capacités de production des usines (en tonnes)
capacite_usines = {
    'Liverpool': 150000,
    'Brighton': 200000
}

# Débit maximal des dépôts (en tonnes)
debit_depots = {
    'Newcastle': 70000,
    'Birmingham': 50000,
    'London': 100000,
    'Exeter': 40000
}

# Demandes des clients (en tonnes)
demande_clients = {
    'C1': 50000,
    'C2': 10000,
    'C3': 40000,
    'C4': 35000,
    'C5': 60000,
    'C6': 20000
}

# Coûts d'expédition (en dollars par tonne)
couts_exp = {
    ('Liverpool', 'Newcastle'): 0.5,
    ('Liverpool', 'Birmingham'): 0.5,
    ('Liverpool', 'London'): 1.0,
    ('Liverpool', 'Exeter'): 0.2,
    ('Brighton', 'Birmingham'): 0.3,
    ('Brighton', 'London'): 0.5,
    ('Brighton', 'Exeter'): 0.2,
    ('Liverpool', 'C1'): 1.0,
    ('Liverpool', 'C3'): 1.5,
    ('Liverpool', 'C4'): 2.0,
    ('Liverpool', 'C6'): 1.0,
    ('Brighton', 'C1'): 2.0,
    ('Newcastle', 'C2'): 1.5,
    ('Newcastle', 'C3'): 0.5,
    ('Newcastle', 'C4'): 1.5,
    ('Birmingham', 'C1'): 1.0,
    ('Birmingham', 'C2'): 0.5,
    ('Birmingham', 'C3'): 0.5,
    ('Birmingham', 'C4'): 1.0,
    ('Birmingham', 'C5'): 0.5,
    ('London', 'C2'): 1.5,
    ('London', 'C3'): 2.0,
    ('Exeter', 'C3'): 0.2,
    ('Exeter', 'C4'): 1.5,
    ('Exeter', 'C5'): 0.5,
    ('Exeter', 'C6'): 1.5
}
```

```python
# Création du modèle d'optimisation
model = pulp.LpProblem("Minimisation_des_Couts_d'Expédition", pulp.LpMinimize)
```

```python
# les variables de décision
X_ud = pulp.LpVariable.dicts("X_ud", [(u, d) for u in capacite_usines for d in debit_depots], lowBound=0, cat='Continuous')
Y_dc = pulp.LpVariable.dicts("Y_dc", [(d, c) for d in debit_depots for c in demande_clients], lowBound=0, cat='Continuous')
Z_uc = pulp.LpVariable.dicts("Z_uc", [(u, c) for u in capacite_usines for c in demande_clients], lowBound=0, cat='Continuous')
```

```python
# Fonction objectif
model += (
    pulp.lpSum([X_ud[(u, d)] * couts_exp[(u, d)] for u in capacite_usines for d in debit_depots if (u, d) in couts_exp]) +
    pulp.lpSum([Y_dc[(d, c)] * couts_exp[(d, c)] for d in debit_depots for c in demande_clients if (d, c) in couts_exp]) +
    pulp.lpSum([Z_uc[(u, c)] * couts_exp[(u, c)] for u in capacite_usines for c in demande_clients if (u, c) in couts_exp])
)
```

```python
# Contraintes de satisfaction de la demande des clients
for c in demande_clients:
    model += (
        pulp.lpSum([Y_dc[(d, c)] for d in debit_depots if (d, c) in couts_exp]) +
        pulp.lpSum([Z_uc[(u, c)] for u in capacite_usines if (u, c) in couts_exp])
    ) == demande_clients[c], f"Satisfaction_Demande_{c}"

# Contraintes de capacité des usines
for u in capacite_usines:
    model += (
        pulp.lpSum([X_ud[(u, d)] for d in debit_depots if (u, d) in couts_exp]) +
        pulp.lpSum([Z_uc[(u, c)] for c in demande_clients if (u, c) in couts_exp])
    ) <= capacite_usines[u], f"Capacite_Usine_{u}"

# Contraintes de débit des dépôts
for d in debit_depots:
    model += (
        pulp.lpSum([Y_dc[(d, c)] for c in demande_clients if (d, c) in couts_exp])
    ) <= debit_depots[d], f"Debit_Depot_{d}"

# Contraintes de flux des dépôts
for d in debit_depots:
    model += (
        pulp.lpSum([X_ud[(u, d)] for u in capacite_usines if (u, d) in couts_exp])
    ) == pulp.lpSum([Y_dc[(d, c)] for c in demande_clients if (d, c) in couts_exp]), f"Flux_Depot_{d}"
```

```python
# Résoudre le problème
model.solve()

# Afficher le statut de la solution
print(f"Status: {pulp.LpStatus[model.status]}")

# Afficher les valeurs optimales des variables de décision
for v in model.variables():
    print(f"{v.name} = {v.varValue}")

# Afficher le coût total minimal
print(f"Coût total minimal d'expédition : {pulp.value(model.objective)} dollars")
```

# Question 2

# Variables de décision

- Xud : Quantité expédiée de l'usine u : (Liverpool, Brighton) au dépôt d : (Newcastle, Birmingham, London, Exeter).
- Ydc : Quantité expédiée du dépôt d : (Newcastle, Birmingham, London, Exeter) au clients c : (c1, c2, c3, c4, c5, c6).
- Zuc : Quantité expédiée de l'usine u : (Newcastle, Birmingham, London, Exeter) au clients c : (c1, c2, c3, c4, c5, c6).
- open_d : Variable binaire qui vaut 1 si le dépôt d est ouvert, 0 sinon.
- augmentation_birmingham : Variable binaire qui vaut 1 si la capacité de Birmingham est étendue, 0 sinon.


# Fonction Objectif

Minimiser le coût total, qui inclut les coûts d'expédition et les coûts d'ouverture des dépôts ainsi que l'extension de Birmingham :

$$
\text{Minimiser } T = 
\sum_{u} 
\sum_{d} 
c_{ud} \cdot X_{ud} 
+ 
\sum_{d} 
\sum_{c} 
c_{dc} \cdot Y_{dc} 
+ 
\sum_{u} 
\sum_{c} 
C_{uc} \cdot Z_{uc}\\ 
+ 
\sum_{d} 
\text{C\_open}_d \cdot \text{open}_d 
+ \text{C\_extension\_birmingham} \cdot \text{extension\_birmingham}
$$


# Contraintes

## Contraintes de satisfaction de la demande des clients :

$$
\sum_{d} Y_{dc} + 
\sum_{u} Z_{uc} = \text{demande}_c, \quad \forall c \in \{\text{C1, C2, C3, C4, C5, C6}\}
$$

## Contraintes de capacité des usines :

$$
\sum_{d} X_{ud} + 
\sum_{c} Z_{uc} \leq \text{capacité}_u, \quad \forall u \in \{\text{Liverpool, Brighton}\}
$$

## Contraintes de débit des dépôts :

$$
\sum_{c} Y_{dc} \leq \text{capacité}_d \cdot \text{open}_d, \quad \forall d \in \{\text{Newcastle, Birmingham, London, Exeter, Bristol, Northampton}\}
$$
Si d = Birmingham, alors la capacité capacité_d inclut l'extension possible :

$$
\text{capacité}_\text{Birmingham} + \text{extension\_birmingham} \times \text{capacity\_birmingham}
$$


## Contraintes de flux des dépôts :

$$
\sum_{u} X_{ud} = 
\sum_{c} Y_{dc}, \quad \forall d \in \{\text{Newcastle, Birmingham, London, Exeter, Bristol, Northampton}\}
$$


## Limitation du nombre de dépôts ouverts :

$$
\sum_{d} \text{open}_d = 4
$$



```python
# les données du problème
usines = ['Liverpool', 'Brighton']
depots = ['Newcastle', 'Birmingham', 'London', 'Exeter', 'Bristol', 'Northampton']
clients = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6']

capacite_usines = {'Liverpool': 150000, 'Brighton': 200000}
capacite_depots = {'Newcastle': 70000, 'Birmingham': 50000, 'London': 100000, 'Exeter': 40000, 'Bristol': 0, 'Northampton': 0}
demande_clients = {'C1': 50000, 'C2': 10000, 'C3': 40000, 'C4': 35000, 'C5': 60000, 'C6': 20000}

couts_exp = {
    ('Liverpool', 'Newcastle'): 0.5,
    ('Liverpool', 'Birmingham'): 0.5,
    ('Liverpool', 'London'): 1.0,
    ('Liverpool', 'Exeter'): 0.2,
    ('Brighton', 'Birmingham'): 0.3,
    ('Brighton', 'London'): 0.5,
    ('Brighton', 'Exeter'): 0.2,
    ('Liverpool', 'C1'): 1.0,
    ('Liverpool', 'C3'): 1.5,
    ('Liverpool', 'C4'): 2.0,
    ('Liverpool', 'C6'): 1.0,
    ('Newcastle', 'C2'): 1.5,
    ('Newcastle', 'C3'): 0.5,
    ('Newcastle', 'C4'): 1.5,
    ('Birmingham', 'C1'): 1.0,
    ('Birmingham', 'C2'): 0.5,
    ('Birmingham', 'C3'): 0.5,
    ('Birmingham', 'C4'): 1.0,
    ('Birmingham', 'C5'): 0.5,
    ('London', 'C2'): 1.5,
    ('London', 'C3'): 2.0,
    ('Exeter', 'C3'): 0.2,
    ('Exeter', 'C4'): 1.5,
    ('Exeter', 'C5'): 0.5,
    ('Exeter', 'C6'): 1.5
}

cout_ouverture = {'Newcastle': 10000, 'Exeter': 5000, 'Bristol': 12000, 'Northampton': 4000}
cout_extension_birmingham = 3000 #$
extension_birmingham = 20000 #tonnes

```

```python
model = pulp.LpProblem("Optimisation_Reseau_Approvisionnement", pulp.LpMinimize)
```

```python
# Variables de décision pour les flux
X_ud = pulp.LpVariable.dicts("X", [(u, d) for u in usines for d in depots], lowBound=0, cat='Continuous')
Y_dc = pulp.LpVariable.dicts("Y", [(d, c) for d in depots for c in clients], lowBound=0, cat='Continuous')
Z_uc = pulp.LpVariable.dicts("Z", [(u, c) for u in usines for c in clients], lowBound=0, cat='Continuous')
```

```python
# Variables binaires pour l'ouverture des dépôts
open_d = pulp.LpVariable.dicts("Open", [d for d in cout_ouverture], cat='Binary')
```

```python
# Variable binaire pour l'extension de Birmingham
extend_birmingham = pulp.LpVariable("Extend_Birmingham", cat='Binary')
```

```python
# Fonction objectif : minimiser le coût total
model += (
    pulp.lpSum(X_ud[(u, d)] * couts_exp[(u, d)] for u in usines for d in depots if (u, d) in couts_exp) +
    pulp.lpSum(Y_dc[(d, c)] * couts_exp[(d, c)] for d in depots for c in clients if (d, c) in couts_exp) +
    pulp.lpSum(Z_uc[(u, c)] * couts_exp[(u, c)] for u in usines for c in clients if (u, c) in couts_exp) +
    pulp.lpSum(open_d[d] * cout_ouverture[d] for d in cout_ouverture) +
    extend_birmingham * cout_extension_birmingham
)
```

```python
# Contraintes de satisfaction de la demande des clients
for i, c in enumerate(clients):
    model += (
        pulp.lpSum(Y_dc[(d, c)] for d in depots if (d, c) in couts_exp) +
        pulp.lpSum(Z_uc[(u, c)] for u in usines if (u, c) in couts_exp) == demande_clients[c]
    ), f"Demande_Client_{c}_{i}"

# Contraintes de capacité des usines
for i, u in enumerate(usines):
    model += (
        pulp.lpSum(X_ud[(u, d)] for d in depots if (u, d) in couts_exp) +
        pulp.lpSum(Z_uc[(u, c)] for c in clients if (u, c) in couts_exp) <= capacite_usines[u]
    ), f"Capacite_Usine_{u}_{i}"

# Contraintes de débit des dépôts (avec extension pour Birmingham)
for i, d in enumerate(depots):
    if d == "Birmingham":
        model += (
            pulp.lpSum(Y_dc[(d, c)] for c in clients if (d, c) in couts_exp) <=
            (capacite_depots[d] + extend_birmingham * extension_birmingham)
        ), f"Capacite_Depot_{d}_{i}"
    elif d in cout_ouverture:
        model += (
            pulp.lpSum(Y_dc[(d, c)] for c in clients if (d, c) in couts_exp) <=
            capacite_depots[d] * open_d[d]
        ), f"Capacite_Depot_{d}_{i}"
    else:
        model += (
            pulp.lpSum(Y_dc[(d, c)] for c in clients if (d, c) in couts_exp) <=
            capacite_depots[d]
        ), f"Capacite_Depot_{d}_{i}"

# Contraintes de flux des dépôts
for d in depots:
    model += (
        pulp.lpSum(X_ud[(u, d)] for u in usines if (u, d) in couts_exp) ==
        pulp.lpSum(Y_dc[(d, c)] for c in clients if (d, c) in couts_exp)
    ), f"Equilibre_Depot_{d}"

# Limitation du nombre de dépôts ouverts
model += (
    pulp.lpSum(open_d[d] for d in cout_ouverture) == 4
), "Limitation_Ouverture_Depots"
```

```python
# Résolution du modèle
model.solve()

# Affichage des résultats
print("Status:", pulp.LpStatus[model.status])
for v in model.variables():
    if v.varValue > 0:
        print(v.name, "=", v.varValue)
```

Il faut étendre la capacité de Birmingham. La variable Extend_Birmingham est égale à 1.0, ce qui signifie que l'extension de la capacité de Birmingham est optimale.

- Birmingham

- Newcastle


En résumé, les dépôts à ouvrir sont Bristol, Exeter, Newcastle et Northampton. Il est optimal d'étendre la capacité de Birmingham. Les dépôts de Birmingham, Exeter et Newcastle seront utilisés pour satisfaire la demande des clients.
